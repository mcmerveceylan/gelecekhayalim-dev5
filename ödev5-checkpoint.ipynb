```python
# Öğrenci notlarını içeren sözlük
notlar = {
    "Merve": {"Matematik": 85, "Fizik": 90, "Kimya": 80},
    "Tugce": {"Matematik": 78, "Fizik": 88, "Kimya": 70},
    "Cansu": {"Matematik": 92, "Fizik": 84, "Kimya": 95},
    "Kubra": {"Matematik": 75, "Fizik": 80, "Kimya": 65},
    "Bahar": {"Matematik": 80, "Fizik": 75, "Kimya": 45}
}

# Kullanıcıdan isim ve ders ismi alma
isim = input("Öğrencinin ismini girin: ")
ders = input("Ders ismini girin (Matematik, Fizik, Kimya): ")

# Notu kontrol etme ve çıktı verme
if isim in notlar:
    if ders in notlar[isim]:
        not_degeri = notlar[isim][ders]
        print("{}'nin {} notu: {}".format(isim, ders, not_degeri))
    else:
        print("Geçersiz ders ismi!")
else:
    print("Öğrenci bulunamadı!")
    
  

while True:
    # Kullanıcıdan işlem seçimi
    print("\nYapmak istediğiniz işlemi seçin:")
    print("1: Notları görüntüle")
    print("2: Not güncelle")
    print("3: Yeni öğrenci ekle")
    print("4: Çıkış")
    
    secim = input("Seçiminizi girin (1-4): ")

    if secim == '1':
        # Notları görüntüleme
        for ogrenci, ders_notlari in notlar.items():
            print(f"{ogrenci}: {ders_notlari}")

    elif secim == '2':
        # Not güncelleme
        isim = input("Güncellemek istediğiniz öğrencinin ismini girin: ")
        if isim in notlar:
            ders = input("Güncellemek istediğiniz ders ismini girin (Matematik, Fizik, Kimya): ")
            if ders in notlar[isim]:
                yeni_not = float(input(f"{ders} notunu girin: "))
                notlar[isim][ders] = yeni_not
                print(f"{isim} için {ders} notu güncellendi: {yeni_not}")
            else:
                print("Geçersiz ders ismi!")
        else:
            print("Öğrenci bulunamadı!")

    elif secim == '3':
        # Yeni öğrenci ekleme
        yeni_isim = input("Yeni öğrencinin ismini girin: ")
        if yeni_isim not in notlar:
            notlar[yeni_isim] = {}
            for ders in ["Matematik", "Fizik", "Kimya"]:
                not_degeri = float(input(f"{yeni_isim} için {ders} notunu girin: "))
                notlar[yeni_isim][ders] = not_degeri
            print(f"{yeni_isim} başarıyla eklendi.")
        else:
            print("Bu öğrenci zaten mevcut!")

    elif secim == '4':
        # Çıkış
         print("Programdan çıkılıyor...")
        break

    else:
        print("Geçersiz seçim, lütfen 1-4 arasında bir sayı girin.")
```
